# Multivariate Analysis

### Popularity of Products in all four locations based on sales

#### calling all individual Regions and their corresponding Sales sum aggregation and assigning it to a variable.

```python
# Sales from the West Region
West_Sales = office_data[office_data['Region'] =='West'].groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
# changing West_Sales columns names 
West_Sales.columns = ['West_Sub-Category', 'West_Sales']
```

```python
West_Sales
```

```python
# Sales from the East Region
East_Sales = office_data[office_data['Region'] =='East'].groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
# changing East_Sales columns names 
East_Sales.columns = ['East_Sub-Category', 'East_Sales']
```

```python
East_Sales
```

```python
# Sales from the Central Region
Central_Sales = office_data[office_data['Region'] =='Central'].groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
# changing Central_Sales columns names 
Central_Sales.columns = ['Central_Sub-Category', 'Central_Sales']
```

```python
Central_Sales
```

```python
# Sales from the South Region
South_Sales = office_data[office_data['Region'] =='South'].groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
# changing South_Sales columns names 
South_Sales.columns = ['South_Sub-Category', 'South_Sales']
```

```python
South_Sales
```

#### merging the above Location-Sales variables assigned

```python
# merging West Sales with East Sales
West_East_Sales = West_Sales.merge(East_Sales, left_index = True, right_index = True, how = 'inner')
```

```python
West_East_Sales
```

```python
# merging West_East Sales with Central Sales
West_East_Central_Sales = West_East_Sales.merge(Central_Sales, left_index = True, right_index = True, how = 'inner')
```

```python
West_East_Central_Sales
```

```python
# merging West_East_Central Sales with South Sales
West_East_Central_South_Sales = West_East_Central_Sales.merge(South_Sales, left_index = True, right_index = True, how = 'inner')
```

```python
West_East_Central_South_Sales
```

### The above shows that in terms of Sales some products are very popular in certain locations than others. Accessories led with huge sales across board. Fasteners recorded the least sales across all regions. Other Products did well in certain locations but not so well in other locations.

## How Discount affected Sales

```python
# checking how many times a particular Discount was utilise in terms of Sales
office_data['Discount'].value_counts()
```

```python
# Aggregative description of Sales
office_data['Sales'].describe()
```

```python
# Graphical(Pie Chart) representation of the relationship between Discount and Sales
office_data.groupby('Discount')['Sales'].sum().plot(kind = 'pie', figsize = (10,10), autopct = '%1.2f%%', fontsize = 12)
plt.title('HOW DISCOUNT AFFECTED SALES', size = 16)
plt.show()
```

## The above shows that even at 0% discount, Sales was higher(47.35%) than that of all other Discounts implemented. A 20% Discount also gave a relatively higher(33.29%) Sales. The above basically shows that some regions did very well in terms of Sales even with a lower Discount implemented while others with huge Discounts(e.g 80%) did comparatively very little in Sales(0.74%)
